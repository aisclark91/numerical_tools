```python
from numpy import linspace
from math import sin
from pylab import plot, show

xpoints = []
ypoints = []

for x in linspace(0,10,100):
    xpoints.append(x)
    ypoints.append(sin(x))

plot(xpoints,ypoints)
show()
```

```python
from pylab import plot,show,ylim, xlabel, ylabel
from numpy import linspace, sin

x = linspace(0,10,100)
y = sin(x)
plot(x,y,'rs') #r,g,b,c,m,y,k,w --> red,green,blue,cyan,magenta,yellow,black,white / -, --, o, s --> solid, dashed, circles, squares
ylim(-1.1,1.1)
xlabel('x')
ylabel('f(x)')
show()
```
