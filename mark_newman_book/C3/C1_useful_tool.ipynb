```python
from numpy import loadtxt
from pylab import plot, show, ylim, xlabel, ylabel

years = []

data = loadtxt('sunspots.txt')
xpoints = data[:,0] #first column: month label from 1749
ypoints = data[:,1] #second column: sunspots

for x in xpoints:
    date =1/12*x+1749
    years.append(date)

plot(years,ypoints)
xlabel('Year')
ylabel('Number of Sunspots')
show()
```

```python
from numpy import loadtxt, array
from pylab import plot, show, ylim, xlabel, ylabel

years = []

data = loadtxt('sunspots.txt',max_rows = 1000)
xpoints = data[:,0] #first column: month label from 1749
ypoints = data[:,1] #second column: sunspots

for x in xpoints:
    date =1/12*x+1749
    years.append(date)

years = array(years)

plot(years,ypoints)
xlabel('Year')
ylabel('Number of Sunspots')
show()
```
